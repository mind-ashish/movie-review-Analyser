```python
#cleaning documents, removing stopwords,stemming using lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ    #adjective
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV         #adverb
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
stops=set(stopwords.words('english'))
```

```python
import string
punctuations=list(string.punctuation)
stops.update(punctuations)
```

```python
stops
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))  #for selecting the type of pos
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clean_review(document),category) for document,category in documents]
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:   #for each word we need to check wheter it appears in a
        #doc or not
        current_features[w]=w in words_set
    return current_features
```

```python
get_feature_dict(training_documents[0][0])  #this shows how each word of feature
#has appeared in this doc or not.
```

```python
training_data=[(get_feature_dict(doc),category) for doc, category in training_documents]
```

```python
testing_data=[(get_feature_dict(doc),category) for doc, category in training_documents]
```
